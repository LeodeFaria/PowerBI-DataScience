# Fundamentos de Python para Ciência de Dados


## 5 - aula: Matplotlib

Matplotlib é uma biblioteca para criação de gráficos e visualização de dados em Python. Durante a exploração dos dados e na avaliação de resultados é muito útil usar gráficos para entender melhor o que está acontecendo. Visualização de dados é por si só uma área grande da ciência de dados, então vamos ver apenas uma introdução com os conceitos mais importantes.

```python
import numpy as np
import pandas as pd
# Só iremos usar o modulo pyplot e iremos chamá-lo de plt
import matplotlib.pyplot as plt
```

Vamos começar com um exemplo simples de um gráfico de linha e a partir dele explorar as funcionalidades do *matplotlib*

```python
# Vamos definir alguns dados ficticios usando listas
empresa_a = [1000, 1500, 1200, 3000, 2100, 2000]
empresa_b = [750, 2300, 1000, 500, 1200, 1500]
```

Para fazer um gráfico de barras usamos a função `plt.plot()`

```python
# Criar o gráfico
plt.plot(empresa_a)
```

```python
# Podemos aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Criar o gráfico
plt.plot(empresa_a);
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Criar o gráfico alterando o estilo
plt.plot(empresa_a, marker='o', linestyle='--', color='green')

# Definir um titulo para o gráfico
plt.title('Lucros')

# Definir a legenda de cada eixo
plt.xlabel('Mês')
plt.ylabel('Valor em R$')
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Criar o gráfico com estilo
plt.plot(empresa_a, 'o--', color='green')

# Definir um titulo para o gráfico
plt.title('Lucros')

# Definir a legenda de cada eixo
plt.xlabel('Mês')
plt.ylabel('Valor em R$')
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Criar os gráfico com estilo e legenda
plt.plot(empresa_a, 'o--', color='green', label='Empresa - A')
plt.plot(empresa_b, 'o--', color='blue',  label='Empresa - B')

# Definir um titulo para o gráfico
plt.title('Lucros')

# Definir a legenda de cada eixo
plt.xlabel('Mês')
plt.ylabel('Valor em R$')

# Indica que queremos colocar a legenda
plt.legend();
```

### Visualizações univariadas

Para essa visualizações iremos utilizar o conjunto de dados do SENADO FEDERAL que contém informações sobre a remuneração dos servidores da casa. Este é o [link](http://www.senado.leg.br/transparencia/rh/servidores/consulta_remuneracao.asp) para acesso.

```python
# Vamos carregar o conjunto de dados que vamos usar
import pandas as pd
df = pd.read_csv('/content/SF_ConsultaRemuneracaoServidoresParlamentares_202103.csv',
                  sep=';', encoding = "ISO-8859-1", decimal = ',')
df.head()
```

#### Gráfico de barra
Primeiro vamos fazer um exemplo simples com dados inventados

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Altura de cada barra
altura = [10, 5, 6, 2, 7]

# Posição de cada barra
posicao = [1, 2, 3, 4, 5]

# Plotando o gráfico
plt.bar(posicao, altura);
```

Agora vamos visualizar o número de servidores de cada espécie de VÍNCULO no Senado Federal.

```python
# .value_counts() conta a frequencia de cada valor
contagem_classes = df['VÍNCULO'].value_counts()
contagem_classes
```

```python
type(contagem_classes)
```

```python
classes = list(contagem_classes.index)
contagens = contagem_classes.values

print(classes)
print(contagens)
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Plotando as barras
plt.bar(classes, contagens);
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Plotando as barras
plt.bar(np.arange(len(classes)), contagens)

# Mudando os valores no eixo x
plt.xticks(np.arange(len(classes)), classes);
```

#### Histograma

Um histograma mostra a distruibuição de uma variável, isso é, a frequencia em que cada uma dos seus valores ocorre. Usando `plt.hist` vamos gerar um histograma para os valores de remuneração líquida dos servidores ('REM_LIQUIDA').

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Basta passarmos a variavel de interesse
plt.hist(df['REM_LIQUIDA']);
```

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(10, 6))

# Podemos mudar o tamanho dos intervalos que são considerados para contagem
plt.hist(df['REM_LIQUIDA'], bins=100);
```

### Visualizações multivariadas

#### Scatterplot

Essa é uma maneira de visualizar a relação entre duas variáveis. Vamos usar a função `plt.scatter` para ver a relação entre o valor da remuneração básica ('REMUN_BASICA') e o valor da remuneração líquida ('REM_LIQUIDA')

Faz sentido estudar corelações entre outras variáveis quantitativas?

```python
# Aumentar o tamanho da imagem
plt.figure(figsize=(15, 8))

# Ploatando os dados
plt.scatter(df['REMUN_BASICA'], df['REM_LIQUIDA'])

# Nomeando os eixos
plt.xlabel('Remuneração básica')
plt.ylabel('Remuneração líquida')
```

Podemos ver que existe uma correlação positiva bem grande entre a renda líquida e básica, mas para valores posítiovs da remuneração básica.
